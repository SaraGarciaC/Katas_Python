1.Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias 
de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def frecuencias_letras (texto):
    """
    Esta función recibe una cadena de texto y devuelve un diccionario con la frecuencia de cada letra.

    Args:
        texto (str): cadena de texto.

    Returns:
        dict[str, int]: diccionario con la frecuencia de cada letra, excluyendo espacios y signos de puntuación.
    """
    
    frecuencias = {}

    for letra in texto:
        if letra.isalpha(): #solo tiene en cuenta letra, excluyendo espacios y signos de puntuación
            letra = letra.lower() # todo en minusculas
            frecuencias[letra] = frecuencias.get(letra,0) +1

    return frecuencias
```

3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe 
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def filtrar_palabras (lista_palabras, palabra_objetivo):
    """
    Esta función toma una lista de palabras y una palabra objetivo como parámetros y devuelve una lista con todas las palabras de la lista
    que contienen la palabra objetivo.

    Args:
        lista_palabras (list[str]): lista de palabras.
        palabra_objetivo (str): palabra objetivo.

    Returns:
        list[str]: lista con todas las palabras que contienen la palabra objetivo.
    """
    lista_filtrada = []

    for palabra in lista_palabras:
        if palabra_objetivo in palabra:
            lista_filtrada.append(palabra)

    return lista_filtrada
```

4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función 
map()

```python
# pedimos al usuario las listas (o podríamos proporcionales directamente). Extraemos valores y convertimos a lista de floats
entrada1 = input("Introduce una lista de numeros separados por comas: ")
lista1 = list(map(float, entrada1.split(",")))

entrada2 = input("Introduce una segunda lista de numeros separados por comas con la misma longitud que la anterior: ")
lista2 = list(map(float, entrada2.split(",")))


def lista_diferencias(lista1, lista2):
    """
    Calcula la diferencia elemento a elemento entre dos listas numéricas de la misma longitud.

    Args:
        lista1 (list[float]): primera lista de números.
        lista2 (list[float]): segunda lista de números (misma longitud que lista1).

    Returns:
        list[float]: lista con las diferencias calculadas como lista1[i] - lista2[i], redondeadas a 2 decimales.

    Raises:
        ValueError: si las listas no tienen la misma longitud.
    """
    
    if len(lista1) != len(lista2):
        raise ValueError("Las listas deben tener la misma longitud")

    return list(map(lambda x, y: round(x - y, 2), lista1, lista2))


try:
    resultado = lista_diferencias(lista1, lista2)
    print(f"La diferencia entre las dos listas es: {resultado}")
except ValueError as e:
    print("Error!", e)


```

5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por 
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual 
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver 
una tupla que contenga la media y el estado.

```python
def evaluar_notas (notas, nota_aprobado=5):
    """
    Esta función evalúa una lista de notas, calcula la media y determina si es un aprobado o un suspenso.

    Args:
        notas (list[float]): lista de números.
        nota_aprobado (float): nota mínima necesaria para aprobar (por defecto, 5).

    Returns:
        tuple[float, str]: tupla (media, resultado), donde resultado es "aprobado" o "suspenso".
    """
    
    media = sum(notas) / len(notas)
    if media >= nota_aprobado:
        resultado = "aprobado"
    else:
        resultado = "suspenso"

    return media, resultado
```

```python
print(evaluar_notas([5,8,3,7,4],5))
print(evaluar_notas ([0,8,4,5]))
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial(n):
    """
Calcula el factorial de un número entero no negativo.

Args:
    n (int): número entero no negativo.

Returns:
    int: factorial de n.

Raises:
    ValueError: si n es negativo.
"""

    if n < 0:
        raise ValueError("El factorial no está definido para números negativos.")

    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)


try:
    entrada = input("Introduce un número entero no negativo. ")
    n = int(entrada)

    resultado = factorial(n)
    print(f"El factorial de {n} es {resultado}")

except ValueError as e:
    print("Error!", e)
```

13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en 
mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función 
map()

```python
def mayusculas_minusculas(caracteres):
    """
    Esta función recibe como argumento un conjunto de caracteres y devuelve una lista de tuplas con cada letra en mayúscula y minúscula,
    comprobando previamente que el conjunto no esté vacío y que sus elementos sean cadenas de texto de un solo carácter (letras).

    Args:
        caracteres (set[str]): conjunto de caracteres (letras).

    Returns:
        list[tuple[str, str]]: lista de tuplas con cada carácter en mayúscula y minúscula.

    Raises:
        ValueError: si el conjunto está vacío o si algún elemento no es una letra.
        TypeError: si algún elemento no es una cadena de texto de un solo carácter.
    """
    if not caracteres:
        raise ValueError("El conjunto está vacío.")

    for c in caracteres:
        if not isinstance(c, str) or len(c) != 1:
            raise TypeError("Los elementos del conjunto deben ser cadenas de texto de un solo carácter.")
        if not c.isalpha():
            raise ValueError(f"El carácter '{c}' no es una letra.")

    return list(map(lambda c: (c.upper(), c.lower()), caracteres))


# set de prueba
caracteres = {"a", "e", "S"}

try:
    print(mayusculas_minusculas(caracteres))
except (TypeError, ValueError) as e:
    print("Error:", e)
```

```python
def palabras_empiezan_por(lista_palabras, letra):
    """
    Esta función recibe como argumento una lista de palabras y devuelve otra lista con aquellas que comienzan con una letra específica.

    Args:
        lista_palabras (list[str]): lista de palabras a filtrar.
        letra (str): letra por la que deben comenzar las palabras.

    Returns:
        list[str]: lista con las palabras filtradas.
    """

    letra = letra.lower()
    return list(filter(lambda x: x.lower().startswith(letra), lista_palabras))

#prueba
letra = "A"
lista_palabras = ["Arbol", "casa", "Analisis", "Rio"]
lista_filtrada = palabras_empiezan_por(lista_palabras,letra)
print(lista_filtrada)
```

```python
import string

def longitud_mayor_a(texto, n):
    """
    Esta función toma un texto cualquiera y un número como argumentos y devuelve una lista
    con las palabras cuya longitud sea mayor que el número especificado.

    Args:
        texto (str): texto cualquiera.
        n (int): número entero positivo.

    Returns:
        list[str]: lista de palabras.
    """
    palabras = texto.split() #para limpiar espacios
    palabras_limpias = [p.strip(string.punctuation) for p in palabras] #para no contabilizar signos de puntutación

    return list(filter(lambda x: len(x) > n, palabras_limpias))

#prueba
texto = "Cuando el grajo vuela bajo, hace un frío del carajo."
palabras_filtradas = longitud_mayor_a(texto, 3)
print(palabras_filtradas)
```

17. Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2]
corresponde al número quinientos setenta y dos [572]. Usa la función reduce()

```python
from functools import reduce

def digitos_a_num(lista_digitos):
    """
    Esta función toma como argumento una lista de dígitos y devuelve el número correspondiente
    usando la función reduce.

    Args:
        lista_digitos (list[int]): lista de dígitos enteros entre 0 y 9.

    Returns:
        int: número entero formado a partir de los dígitos.

    Raises:
        ValueError: si la lista está vacía o contiene valores que no son dígitos válidos.
    """
    if not lista_digitos:
        raise ValueError("La lista no puede estar vacía.")
    
    for d in lista_digitos:
        if not isinstance(d, int) or d < 0 or d > 9:
            raise ValueError("La lista debe estar compuesta por números enteros entre 0 y 9")
    
    return reduce(lambda numero, digito: numero * 10 + digito, lista_digitos, 0)

try:
    lista_digitos = [8,7,2] #lista de prueba
    print(digitos_a_num(lista_digitos))

except ValueError as e:
    print("Error!", e)
```

22. Dada una lista numérica, obtén el producto total de los valores de dicha lista.Usa la función reduce() .

```python
from functools import reduce

def prod_tot(lista_valores):
    """
    Esta función toma como argumento una lista de números y devuelve el producto total
    de todos los valores de la lista.

    Args:
        lista_valores (list[float | int]): lista de valores numéricos.

    Returns:
        float | int: el producto de los números de la lista.

    Raises:
        TypeError: si algún elemento no es numérico.
        ValueError: si la lista está vacía.
    """
    if not lista_valores:
        raise ValueError("La lista no puede estar vacía")

    for n in lista_valores:
        if not isinstance(n, (int, float)):
            raise TypeError("Todos los elementos de la lista deben ser números (int o float)")

    return reduce(lambda x, y: x * y, lista_valores, 1)


try:
    lista_valores = [8,15,45] #lista de prueba
    print(prod_tot(lista_valores))
except TypeError as e:
    print("Error!",e)
```

23. Concatena una lista de palabras.Usa la función reduce() .

```python
from functools import reduce
   
def concatenar_palabras(lista_palabras):
    """
    Esta función recibe una lista de palabras y devuelve una cadena con todas ellas
    concatenadas en una frase legible, separadas por espacios.

    Args:
        lista_palabras (list[str]): lista de palabras a concatenar.

    Returns:
        str: cadena resultante con las palabras concatenadas y separadas por espacios.
    """
    
    if not lista_palabras:
        return ""

    return reduce(lambda a, b: a + " " + b, lista_palabras)


lista_palabras = ["A", "mal", "tiempo", "buena", "cara"]
print(concatenar_palabras(lista_palabras))
```

24. Calcula la diferencia total en los valores de una lista. Usa la función reduce() .

```python
from functools import reduce

def dif_total(lista_val):
    """
    Esta función recibe una lista de valores numéricos (int o float) y devuelve la diferencia total
    acumulada entre los valores de la lista.

    Args:
        lista_val (list[float | int]): lista de valores numéricos.

    Returns:
        float | int: resultado de restar todos los valores de la lista de forma acumulada.

    Raises:
        ValueError: si la lista está vacía.
        TypeError: si algún elemento no es numérico.
        
    """
    if not lista_val:
        raise ValueError("La lista no puede estar vacía")
    
    for n in lista_val:
        if not isinstance(n, (int, float)):
            raise TypeError("Todos los elementos de la lista deben ser números (int o float)")

    return reduce(lambda a, b: a - b, lista_val)


# Prueba
try:
    lista_val = [100, 20, 5]
    print(dif_total(lista_val))  # ((100 - 20) - 5) = 75
except (TypeError, ValueError) as e:
    print("Error!", e)
```

26. Crea una función lambda que calcule el resto de la división entre dos números dados.

```python
def calcular_resto(a, b):
    """
    Esta función recibe como argumentos dos números y devuelve el resto de la división entre ambos
    usando una función lambda.

    Args:
        a (int | float): primer número.
        b (int | float): segundo número (divisor).

    Returns:
        int | float: resto de la división de a entre b.

    Raises:
        TypeError: si alguno de los argumentos no es numérico.
        ZeroDivisionError: si b es 0.
    """
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Los valores introducidos deben ser numéricos (int o float).")
    
    if b == 0:
        raise ZeroDivisionError("No se puede dividir entre 0.")
    
    resto = lambda x, y: x % y
    return resto(a, b)


try:
    print(calcular_resto(18,0)) #prueba
except (TypeError, ZeroDivisionError) as e:
    print ("Error!", e)
```

```python
def calcular_promedio(numeros):
    """
    Esta función recibe una lista de números y devuelve su promedio.

    Args:
        numeros (list[float | int]): lista de números.

    Returns:
        float: promedio de los números de la lista.

    Raises:
        TypeError: si algún elemento no es numérico.
        ValueError: si la lista está vacía.

    """
    
    if not isinstance(numeros, list):
        raise TypeError("Debes introducir una lista de números.")

    if not numeros:
        raise ValueError("La lista no puede estar vacía.")

    for n in numeros:
        if not isinstance(n, (int, float)):
            raise TypeError("Todos los elementos de la lista deben ser números (int o float).")

    return sum(numeros) / len(numeros)


#prueba
try:
    lista = [8, 9, 55, 44, 23]
    print(f"El promedio de los numeros de la lista es {calcular_promedio(lista)}")
except (TypeError, ValueError) as e:
    print("Error!", e)
```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras 
pero en diferente orden.

```python
def son_anagramas(palabra1, palabra2):
    """
    Esta función recibe dos palabras y determina si son anagramas, es decir,
    si están formadas por las mismas letras en diferente orden.

    Args:
        palabra1 (str): primera palabra.
        palabra2 (str): segunda palabra.

    Returns:
        bool: True si son anagramas, False en caso contrario.

    Raises:
        TypeError: si alguno de los argumentos no es una cadena de texto.
    """
    
    if not isinstance(palabra1, str) or not isinstance(palabra2, str):
        raise TypeError("Ambos argumentos deben ser cadenas de texto.")

    # Normalización
    p1 = palabra1.replace(" ", "").lower()
    p2 = palabra2.replace(" ", "").lower()

    return sorted(p1) == sorted(p2)
```

34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: 
crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . 
El objetivo es implementar estos métodos para manipular la estructura del árbol.

Código a seguir:

    1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
    2. Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
    3. Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
    4. Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.

    6. Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes 
    de las mismas.

Caso de uso:

    1. Crear un árbol.
    2. Hacer crecer el tronco del árbol una unidad.
    3. Añadir una nueva rama al árbol.
    4. Hacer crecer todas las ramas del árbol una unidad.
    5. Añadir dos nuevas ramas al árbol.
    6. Retirar la rama situada en la posición 2.
    7. Obtener información sobre el árbol

```python
# Codigo
# 1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
class Arbol: 
    """
    Representa un árbol genérico con un tronco y una lista de ramas.

    El árbol comienza con un tronco de longitud 1 y sin ramas. Proporciona métodos
    para hacer crecer el tronco, añadir y modificar ramas, eliminar ramas y obtener
    información sobre su estado.

    """
    def __init__(self):
        """
        Inicializa un árbol con un tronco de longitud 1 y una lista vacía de ramas.
        
        """
        self.tronco = 1
        self.ramas = []

# 2. Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
    def crecer_tronco(self):
        """
        Incrementa en una unidad la longitud del tronco del árbol.

        """
        self.tronco += 1

# 3. Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
    def nueva_rama(self):
        """
        Añade una nueva rama al árbol con longitud inicial 1.

        """
        self.ramas.append(1)

# 4. Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
    def crecer_ramas(self):
        """ 
        Incrementa en una unidad la longitud de todas las ramas existentes.

        """
        self.ramas = [rama + 1 for rama in self.ramas]
        
# 5. Implementar el método quitar_rama para eliminar una rama en una posición específica.
    def quitar_rama(self, posicion):
        """
        Elimina la rama situada en la posición indicada.

        Args:
            posicion (int): Índice (base 0) de la rama que se desea eliminar.

        Raises:
            TypeError: Si la posición no es un entero.
            IndexError: Si la posición está fuera del rango de la lista de ramas.
            
        """

        if not isinstance (posicion, int):
            raise TypeError ("La posicion debe ser un número entero.")
        
        if posicion < 0 or posicion >= len(self.ramas):
            raise IndexError ("Posición fuera de rango.")

        self.ramas.pop(posicion)

# 6. Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.
    def info_arbol(self):
        """
        Devuelve información sobre el estado actual del árbol.

        La información incluye la longitud del tronco, el número de ramas y una lista
        con las longitudes de cada rama.

        Returns:
            dict: Diccionario con las claves:
                - 'longitud_tronco' (int): Longitud del tronco.
                - 'numero_ramas' (int): Número total de ramas.
                - 'longitud_ramas' (list): Lista con las longitudes de las ramas.
                
        """
        
        return {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitud_ramas": self.ramas.copy() #usamos copy para no modificar el objeto por error
        }
# Caso de uso
# 1. Crear un arbol.
abeto = Arbol()

# 2. Hacer crecer el tronco del árbol una unidad.
abeto.crecer_tronco()

# 3. Añadir una nueva rama al árbol.
abeto.nueva_rama()

# 4. Hacer crecer todas las ramas del árbol una unidad.
abeto.crecer_ramas()

# 5.Añadir dos nuevas ramas al árbol.
abeto.nueva_rama()
abeto.nueva_rama()

# 6. Retirar la rama situada en la posición 2.
abeto.quitar_rama(1)

#7. Obtener información sobre el árbol.
print(abeto.info_arbol())
    
```

```python
class UsuarioBanco:
    """
    Representa a un usuario de un banco con un nombre, un saldo y si tiene o no cuenta corriente.
    Permite realizar operaciones como retirar dinero, transferir dinero y agregar dinero al saldo.
    """

    # 1. Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente
    def __init__(self, nombre, saldo, cuenta_corriente):
        """
        Inicializa un usuario del banco.

        Args:
            nombre (str): nombre del usuario.
            saldo (float | int): saldo inicial del usuario.
            cuenta_corriente (bool): indica si el usuario tiene cuenta corriente.
        """
        if not isinstance(nombre, str):
            raise TypeError("El nombre tiene que ser una cadena de texto (string).")

        if not isinstance(saldo, (int, float)):
            raise TypeError("El saldo tiene que ser un número (int o float).")

        if saldo < 0:
            raise ValueError("No se permite un saldo negativo.")

        if not isinstance(cuenta_corriente, bool):
            raise TypeError("La cuenta corriente ha de ser un bool (True o False).")

        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    # 2. Implementar el método retirar_dinero
    def retirar_dinero(self, cantidad):
        """
        Retira una cantidad de dinero del saldo del usuario.

        Args:
            cantidad (float | int): cantidad a retirar.

        Raises:
            ValueError: si la cantidad es negativa o si no hay saldo suficiente.
            TypeError: si la cantidad no es numérica.
        """
        if not isinstance(cantidad, (int, float)):
            raise TypeError("La cantidad a retirar tiene que ser un número (int o float).")

        if cantidad <= 0:
            raise ValueError("Introduce un importe superior a 0.")

        if cantidad > self.saldo:
            raise ValueError(
                f"No hay fondos suficientes. Elige una cantidad inferior o igual a {self.saldo}."
            )

        self.saldo -= cantidad

    # 3. Implementar el método transferir_dinero
    def transferir_dinero(self, otro_usuario, cantidad):
        """
        Realiza una transferencia de dinero desde otro usuario al usuario actual.

        Args:
            otro_usuario (UsuarioBanco): usuario desde el que se transfiere el dinero.
            cantidad (float | int): cantidad a transferir.

        Raises:
            TypeError: si otro_usuario no es un UsuarioBanco o si cantidad no es numérica.
            ValueError: si no hay saldo suficiente o si alguno no tiene cuenta corriente.
        """
        if not isinstance(otro_usuario, UsuarioBanco):
            raise TypeError("La transferencia solo puede realizarse desde otro UsuarioBanco.")

        if not isinstance(cantidad, (int, float)):
            raise TypeError("Introduce un importe numérico para transferir.")

        if cantidad <= 0:
            raise ValueError("Introduce un importe válido (mayor que 0).")

        if not self.cuenta_corriente or not otro_usuario.cuenta_corriente:
            raise ValueError(
                "Ambos usuarios deben tener cuenta corriente para realizar la transferencia."
            )

        if cantidad > otro_usuario.saldo:
            raise ValueError(
                f"Saldo insuficiente. Introduce un importe inferior o igual a {otro_usuario.saldo}."
            )

        otro_usuario.saldo -= cantidad
        self.saldo += cantidad

    # 4. Implementar el método agregar_dinero
    def agregar_dinero(self, cantidad):
        """
        Agrega una cantidad de dinero al saldo del usuario.

        Args:
            cantidad (float | int): cantidad a agregar.

        Raises:
            ValueError: si la cantidad es negativa o cero.
            TypeError: si la cantidad no es numérica.
        """
        if not isinstance(cantidad, (int, float)):
            raise TypeError("Introduce un valor numérico para agregar.")

        if cantidad <= 0:
            raise ValueError("Introduce un importe válido. Ha de ser mayor que 0.")

        self.saldo += cantidad


# Caso de uso
try:
    # 1. Crear dos usuarios
    alicia = UsuarioBanco("Alicia", 100, True)
    bob = UsuarioBanco("Bob", 50, True)

    # 2. Agregar 20 unidades al saldo de Bob
    bob.agregar_dinero(20)

    # 3. Hacer una transferencia de 80 unidades desde Bob a Alicia
    alicia.transferir_dinero(bob, 80)

    # 4. Retirar 50 unidades de saldo a Alicia
    alicia.retirar_dinero(50)

    # Comprobar el estado final de las dos cuentas
    print("El saldo actualizado de Alicia es:", alicia.saldo)
    print("El saldo actualizado de Bob es:", bob.saldo)

except (ValueError, TypeError) as e:
    print("Error:", e)

# Arrojará error porque Bob no tiene suficiente saldo para transferir 80.
# Tras agregar 20, Bob tendría 70.
  
```

    número de argumentos variable según la opción indicada.

Caso de uso:
    Comprueba el funcionamiento completo de la función procesar_texto


```python
import string

# 1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.

def contar_palabras(texto):
    """
    Cuenta cuántas veces aparece cada palabra en el texto.

    Args:
        texto (str): Texto a analizar.

    Returns:
        dict: Diccionario {palabra: frecuencia}.
    """
    if not isinstance(texto, str):
        raise TypeError("El texto tiene que ser una cadena (str)")

    # Pasamos a minúsculas, separamos y limpiamos signos de puntuación
    palabras = [p.strip(string.punctuation).lower() for p in texto.split() if p.strip(string.punctuation)]

    conteo = {}
    for p in palabras:
        conteo[p] = conteo.get(p, 0) + 1
    return conteo


# 2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva. Tiene que devolver el texto con el remplazo de palabras.
def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """
    Reemplaza palabra_original por palabra_nueva en el texto.

    Args:
        texto (str): Texto original.
        palabra_original (str): Palabra a reemplazar.
        palabra_nueva (str): Palabra nueva.

    Returns:
        str: Texto con el reemplazo aplicado.
    """
    if not all(isinstance(valor, str) for valor in (texto, palabra_original, palabra_nueva)):
        raise TypeError("El texto original, la palabra original y la palabra nueva deben de ser strings.")

    return texto.replace(palabra_original, palabra_nueva)


# 3. Crear funcion eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
def eliminar_palabra(texto, palabra):
    """
    Elimina una palabra del texto.

    Args:
        texto (str): Texto original.
        palabra (str): Palabra a eliminar.

    Returns:
        str: Texto con la palabra eliminada.
    """
    if not all(isinstance(x, str) for x in (texto, palabra)):
        raise TypeError("El texto y la palabra a eliminar han de ser de tipo string")

    palabras = texto.split()  # separamos el texto en una lista de palabras

    # Eliminamos la palabra ignorando puntuación y mayúsculas/minúsculas. 
    filtradas = [p for p in palabras if p.strip(string.punctuation).lower() != palabra.lower()] #devuelve nueva lista sin la palabra eliminada

    return " ".join(filtradas) #devuelve un texto nuevo sin la palabra eliminada y separando strings con espacios


# 4. Crear la función procesar_texto que tome un texto, una opción entre ("contar", "reemplazar", "eliminar")  y un número de argumentos variable según la opción indicada.
def procesar_texto(texto, opcion, *args):
    """
    Procesa un texto según la opción indicada: 'contar', 'reemplazar' o 'eliminar'.

    Args:
        texto (str): Texto a procesar.
        opcion (str): Una de estas opciones: 'contar', 'reemplazar', 'eliminar'.
        *args: Argumentos extra según la opción:
            - 'contar' -> sin args
            - 'reemplazar' -> (palabra_original, palabra_nueva)
            - 'eliminar' -> (palabra)

    Returns:
        dict | str: Resultado del procesamiento.

    Raises:
        TypeError: si los tipos de los argumentos no son correctos.
        ValueError: si la opción no es válida o faltan argumentos.

    """
    if not isinstance(texto, str):
        raise TypeError("El texto a procesar ha de ser tipo string.")

    if not isinstance(opcion, str):
        raise TypeError("La opción debe ser de tipo string.")

    opcion = opcion.lower().strip()

    if opcion == "contar":
        return contar_palabras(texto)

    elif opcion == "reemplazar":
        if len(args) != 2:
            raise ValueError("Para 'reemplazar' hacen falta dos argumentos: palabra_original y palabra_nueva.")
        return reemplazar_palabras(texto, args[0], args[1])

    elif opcion == "eliminar":
        if len(args) != 1:
            raise ValueError("Para 'eliminar' hace falta un argumento: palabra_eliminar.")
        return eliminar_palabra(texto, args[0])

    else:
        raise ValueError('Opción no válida: elige una entre "contar", "reemplazar" y "eliminar".')


# Caso de uso: Comprobar el funcionamiento completo de la función procesar_texto

texto = """El ayuntamiento de Villacódigo informó ayer de que un gato programador fue visto trabajando en una biblioteca pública.
El gato programador, según testigos, tecleaba con mucha concentración mientras que otro gato miraba la pantalla.
Los vecinos dicen que el gato programador pidió una pizza, pero la pizza nunca llegó porque el repartidor se perdió.
Aun así, el gato programador siguió programando y declaró que la programación es mejor que la siesta."""

print(procesar_texto(texto, "contar"))
print(procesar_texto(texto, "reemplazar", "gato", "perro"))
print(procesar_texto(texto, "eliminar", "pizza"))

```

38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.

```python
def momento_del_dia(hora_input):
    """
    Determina si es mañana (día), tarde o noche según la hora proporcionada.

    Args:
        hora_input (str): hora en formato "HH:MM"

    Returns:
        str: "Día", "Tarde" o "Noche"
    """

    try:
        hora, minuto = hora_input.split(":")
        hora = int(hora)
        minuto = int(minuto)
    except:
        return "Formato de hora no válido. Usa HH:MM"

    if not (0 <= hora < 24 and 0 <= minuto < 60):
        return "Hora no válida."

    if 6 <= hora < 12:
        return "Día"
    elif 12 <= hora < 20:
        return "Tarde"
    else:
        return "Noche"
```

40.  Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o  "triángulo") y
datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math

def area_figura(figura, datos):
    """
    Calcula el área de una figura geométrica.

    Args:
        figura (str): Tipo de figura: "rectangulo", "circulo" o "triangulo".
        datos (tuple): Tupla con los datos necesarios:
            - rectangulo: (base, altura)
            - circulo: (radio,)
            - triangulo: (base, altura)

    Returns:
        float: Área de la figura.

    Raises:
        TypeError: Si los tipos de datos no son correctos.
        ValueError: Si la figura no es válida o los datos son incorrectos.

    """
    
    if not isinstance(figura, str):
        raise TypeError("La figura ha de ser una cadena de texto.")
    
    if not isinstance(datos, tuple):
        raise TypeError("Los datos de la figura se deben introducir mediante una tupla.")

    figura = figura.lower()

    if figura not in ("rectangulo", "circulo", "triangulo"):
        raise ValueError('Figura no válida. Elige una entre "rectangulo", "circulo" o "triangulo".')

    # Validamos que los datos sean numéricos y positivos
    for d in datos:
        if not isinstance(d, (int, float)):
            raise TypeError("Los datos deben ser numéricos.")
        if d <= 0:
            raise ValueError("Los valores numéricos de los datos deben ser positivos.")

    if figura == "rectangulo":
        if len(datos) != 2:
            raise ValueError("Para calcular el área de un rectángulo se necesitan dos parámetros. Introduce una tupla con: (base, altura).")
        base, altura = datos
        return round(base * altura, 2)

    elif figura == "circulo":
        if len(datos) != 1:
            raise ValueError("Para calcular el área de un círculo se necesita un parámetro. Introduce una tupla con: (radio,).")
        (radio,) = datos
        return round(math.pi * radio**2, 2)

    else:  # triangulo
        if len(datos) != 2:
            raise ValueError("Para calcular el área de un triángulo se necesitan dos parámetros. Introduce una tupla con: (base, altura).")
        base, altura = datos
        return round((base * altura) / 2, 2)


# Pruebas
try:
    print("Área rectángulo:", area_figura("rectangulo", (58, 6)))
    print("Área círculo:", area_figura("circulo", (5,)))
    print("Área triángulo:", area_figura("triangulo", (4, 6)))
except (TypeError, ValueError) as e:
    print("Error:", e)

        
```
